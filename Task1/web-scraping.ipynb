## Scraping 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd 
```

```python
text = []
aircraft = []
traveller_type = []
seat_type = []
flown_date = []
seat_comortable = []
cabin_staff_service = []
food = []
ground_service = []
value_for_money = []
recommended = []
```

```python
page_size = 100 
page_numbers = 102
```

```python
try:
    web = webdriver.Chrome()

    for i in range(48, page_numbers+1):
        print(f'Scraping Page -> {i}')
        web.get(f'https://www.airlinequality.com/airline-reviews/british-airways/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}')
        web.minimize_window()
        # Wait for the page to load
        # time.sleep(5)  # or use WebDriverWait for better control
    
        # find all texts
        text_elements = web.find_elements(By.CLASS_NAME, 'text_content')
        for text_ in text_elements:
            text.append(text_.text.split('|')[-1].strip())
        
        # Find all divs with the class 'review-stats'
        table_div = web.find_elements(By.CLASS_NAME, 'review-stats')
        
        for div in table_div:
            tr_elements = div.find_elements(By.TAG_NAME, 'tr')
            aircraft_found = False  # Flag to check if "Aircraft" row was found
            traveller_type_found = False 
            seat_type_found = False
            flown_date_found = False
            seat_comortable_found = False 
            cabin_staff_service_found = False
            food_found = False 
            ground_service_found = False 
            value_for_money_found = False 
            recommended_found = False
            
            for tr in tr_elements:
                # aircraft elements 
                aircraft_td = tr.find_element(By.CLASS_NAME, 'review-rating-header')
                # Check if this is the "Aircraft" header
                if 'aircraft' in aircraft_td.get_attribute('class') and aircraft_td.text.strip() == 'Aircraft':
                    aircraft_element = tr.find_element(By.CLASS_NAME, 'review-value')
                    # print(f"Aircraft: {aircraft_element.text}")
                    aircraft.append(aircraft_element.text)
                    aircraft_found = True  # Set the flag to True if "Aircraft" is found
                
                # traveller type elements
                traveller_td = tr.find_element(By.CLASS_NAME, 'review-rating-header')
                # Print out the class name and text for debugging
                class_attr = traveller_td.get_attribute('class')
                text_attr = traveller_td.text.strip()
                # print(f"Class: {class_attr}, Text: {text_attr}")
                
                if 'type_of_traveller' in class_attr and text_attr == 'Type Of Traveller':
                    traveller_element = tr.find_element(By.CLASS_NAME, 'review-value')
                    # print(f"Traveller Type: {traveller_element.text}")
                    traveller_type.append(traveller_element.text)
                    traveller_type_found = True
               
    
                # seat type elements
                seat_td = tr.find_element(By.CLASS_NAME, 'review-rating-header')
                # Print out the class name and text for debugging
                class_attr = seat_td.get_attribute('class')
                text_attr = seat_td.text.strip()
                # print(f"Class: {class_attr}, Text: {text_attr}")
                
                if 'cabin_flown ' in class_attr and text_attr == 'Seat Type':
                    seat_element = tr.find_element(By.CLASS_NAME, 'review-value')
                    # print(f"seat_element Type: {seat_element.text}")
                    seat_type.append(seat_element.text)
                    seat_type_found = True
                    
                # flown date elements
                date_td = tr.find_element(By.CLASS_NAME, 'review-rating-header')
                # Print out the class name and text for debugging
                class_attr = date_td.get_attribute('class')
                text_attr = date_td.text.strip()
                # print(f"Class: {class_attr}, Text: {text_attr}")
                if 'date_flown' in class_attr and text_attr == 'Date Flown':
                    date_element = tr.find_element(By.CLASS_NAME, 'review-value')
                    # print(f"date_element Type: {date_element.text}")
                    flown_date.append(date_element.text)
                    flown_date_found = True
                    
                # seat comortable elements
                seat_comftr_td = tr.find_element(By.CLASS_NAME, 'review-rating-header')
                # Print out the class name and text for debugging
                class_attr = seat_comftr_td.get_attribute('class')
                text_attr = seat_comftr_td.text.strip()
                # print(f"Class: {class_attr}, Text: {text_attr}")
                
                if 'seat_comfort' in class_attr and text_attr == 'Seat Comfort':
                    seat_comftr_element = tr.find_element(By.CLASS_NAME, 'review-rating-stars')
                    stars = len(seat_comftr_element.find_elements(By.CLASS_NAME, 'star.fill'))
                    # print(f"seat_comftr_element: {seat_comftr_element}")
                    # print(f"# seat_comft stars : {stars}")
    
                    seat_comortable.append(stars)
                    seat_comortable_found = True
                
                # Cabin Staff Service elements
                cabin_serv_td = tr.find_element(By.CLASS_NAME, 'review-rating-header')
                # Print out the class name and text for debugging
                class_attr = cabin_serv_td.get_attribute('class')
                text_attr = cabin_serv_td.text.strip()
                # print(f"Class: {class_attr}, Text: {text_attr}")
                
                if 'cabin_staff_service' in class_attr and text_attr == 'Cabin Staff Service':
                    cabin_serv_element = tr.find_element(By.CLASS_NAME, 'review-rating-stars')
                    stars = len(cabin_serv_element.find_elements(By.CLASS_NAME, 'star.fill'))
                    # print(f"seat_comftr_element: {seat_comftr_element}")
                    # print(f"# cabin stars : {stars}")
    
                    cabin_staff_service.append(stars)
                    cabin_staff_service_found = True
               
                # Food & Beverages elements
                food_td = tr.find_element(By.CLASS_NAME, 'review-rating-header')
                # Print out the class name and text for debugging
                class_attr = food_td.get_attribute('class')
                text_attr = food_td.text.strip()
                # print(f"Class: {class_attr}, Text: {text_attr}")
                
                if 'food_and_beverages' in class_attr and text_attr == 'Food & Beverages':
                    food_element = tr.find_element(By.CLASS_NAME, 'review-rating-stars')
                    stars = len(food_element.find_elements(By.CLASS_NAME, 'star.fill'))
                    # print(f"seat_comftr_element: {seat_comftr_element}")
                    # print(f"# food stars: {stars}")
    
                    food.append(stars)
                    food_found = True
                    
                # Ground Service elements
                ground_serv_td = tr.find_element(By.CLASS_NAME, 'review-rating-header')
                # Print out the class name and text for debugging
                class_attr = ground_serv_td.get_attribute('class')
                text_attr = ground_serv_td.text.strip()
                # print(f"Class: {class_attr}, Text: {text_attr}")
                
                if 'ground_service' in class_attr and text_attr == 'Ground Service':
                    ground_serv_element = tr.find_element(By.CLASS_NAME, 'review-rating-stars')
                    stars = len(ground_serv_element.find_elements(By.CLASS_NAME, 'star.fill'))
                    # print(f"seat_comftr_element: {seat_comftr_element}")
                    # print(f"# ground stars : {stars}")
    
                    ground_service.append(stars)
                    ground_service_found = True
                
                # Value For Money elements
                money_value_td = tr.find_element(By.CLASS_NAME, 'review-rating-header')
                # Print out the class name and text for debugging
                class_attr = money_value_td.get_attribute('class')
                text_attr = money_value_td.text.strip()
                # print(f"Class: {class_attr}, Text: {text_attr}")
                
                if 'value_for_money' in class_attr and text_attr == 'Value For Money':
                    money_value_element = tr.find_element(By.CLASS_NAME, 'review-rating-stars')
                    stars = len(money_value_element.find_elements(By.CLASS_NAME, 'star.fill'))
                    # print(f"seat_comftr_element: {seat_comftr_element}")
                    # print(f"# money_value stars : {stars}")
    
                    value_for_money.append(stars)
                    value_for_money_found = True
            
                # Recommended elements
                recommended_td = tr.find_element(By.CLASS_NAME, 'review-rating-header')
                # Print out the class name and text for debugging
                class_attr = recommended_td.get_attribute('class')
                text_attr = recommended_td.text.strip()
                # print(f"Class: {class_attr}, Text: {text_attr}")
                
                if 'recommended' in class_attr and text_attr == 'Recommended':
                    recommended_element = tr.find_element(By.CLASS_NAME, 'review-value')
                    # print(f"seat_comftr_element: {seat_comftr_element}")
                    # print(f"Recommended: {recommended_element.text}")
    
                    recommended.append(recommended_element.text)
                    recommended_found = True
                
                
                
            if not aircraft_found:
                # print("None")  # Print None if no "Aircraft" row was found
                aircraft.append('None')
            if not traveller_type_found:
                # print("None")  # Print None if no "Aircraft" row was found
                traveller_type.append('None')
            
              
            if not seat_type_found:
                # print("None")  # Print None if no "Aircraft" row was found
                seat_type.append('None')
    
            if not flown_date_found:
                # print("None")  # Print None if no "Aircraft" row was found
                flown_date.append('None')
    
            if not seat_comortable_found:
                # print("None")  # Print None if no "Aircraft" row was found
                seat_comortable.append('None')
    
            if not cabin_staff_service_found:
                # print("None")  # Print None if no "Aircraft" row was found
                cabin_staff_service.append('None')
    
            if not food_found:
                # print("None")  # Print None if no "Aircraft" row was found
                food.append('None')
            
            if not ground_service_found:
                # print("None")  # Print None if no "Aircraft" row was found
                ground_service.append('None')
    
            if not value_for_money_found:
                # print("None")  # Print None if no "Aircraft" row was found
                value_for_money.append('None')
            
            if not recommended_found:
                # print("None")  # Print None if no "Aircraft" row was found
                recommended.append('None')
            
        # number of reviews was scrapped 
        print(f'    total reviews -----> {len(text)}')

    web.quit()

except Exception as e:
    print(f"Maybe something is wrong, bro. Error: {e}")
```

## Save Scrapped Date in CSV file 

```python
#Combine lists into a dictionary (keys will be column names)
data = {
    'text': text,
    'aircraft': aircraft,
    'traveller_type': traveller_type,
    'seat_type': seat_type,
    'flown_date': flown_date,
    'seat_comortable': seat_comortable,
    'cabin_staff_service': cabin_staff_service,
    'food': food,
    'ground_service': ground_service,
    'value_for_money': value_for_money,
    'recommended': recommended
}


# Create DataFrame
df = pd.DataFrame(data)

# Save DataFrame to CSV file
df.to_csv('british-airways-reviews.csv', index=False)
```
